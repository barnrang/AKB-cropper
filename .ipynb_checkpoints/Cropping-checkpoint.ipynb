```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('images/sample2.jpg',0)
```

```python
print(img.shape)
plt.imshow(img, cmap='gray')
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,10)
plt.imshow(th2, cmap='gray')
```

```python
th3 = cv2.bitwise_not(th2)
plt.imshow(th3, cmap='gray')
```

```python
kernel = np.ones((5,5),np.uint8)
horizontalsize = np.int(th3.shape[0]/30.)
horizontalstructure = cv2.getStructuringElement(cv2.MORPH_RECT, (horizontalsize, 1))
erode = cv2.erode(th3,horizontalstructure,iterations = 1)
dilate_h = cv2.dilate(erode, horizontalstructure, iterations = 1)
plt.imshow(dilate_h)
```

```python
kernel = np.ones((3,3),np.uint8)
verticalsize = np.int(th3.shape[1]/50)
verticalstructure = cv2.getStructuringElement(cv2.MORPH_RECT, (1,verticalsize))
erode = cv2.erode(th3,verticalstructure,iterations = 1)
dilate_v = cv2.dilate(erode, verticalstructure, iterations = 3)
plt.imshow(dilate_v)
```

```python
cv2.merge((dilate_h, dilate_v)).shape
```

```python
kernel = np.ones((3,3),np.uint8)
dilation = cv2.dilate(dilate_v,kernel,iterations = 1)
plt.imshow(dilation)
```

```python
kernel = np.ones((3,3),np.uint8)
dilation = cv2.dilate(th3,kernel,iterations = 1)
plt.imshow(dilation)
```

```python
image = img.copy()
lines = cv2.HoughLines(dilation,1,np.pi/180,100)
for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    cv2.line(image,(x1,y1),(x2,y2),(0,0,255),2)
    
plt.imshow(image)
```
